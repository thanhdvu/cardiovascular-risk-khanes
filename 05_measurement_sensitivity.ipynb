# 05. 측정 방식 변화 검토 및 민감도 분석

**교수님 피드백 반영 — 보조 민감도 분석 노트북**

지난 메일에서 주신 방향을 바탕으로, COVID 시기별 유병률 감소가 실제 변화인지 측정 방식 변화에 따른 인공적인 변화인지 검토하고, 후속 분석 방향을 정리한다.

검토 항목(교수님 피드백):
1. 혈압: **원자료 분석 vs 연도별 측정기기 변화 보정 vs 측정 일관 연도 제한분석** 을 함께 제시
2. 추가 변수: **결측률 ≤30% → 결측치 대체 후 사용 / >30% → 변수 삭제**, HDL·LDL은 측정 시기별 변화 주의
3. `covid_period` 는 측정 장비/환경 변화와 분리되기 어려우므로 **인과 해석 불가**, 민감도분석으로 제시

```python
import pyreadstat, pandas as pd, numpy as np, glob, warnings
from pathlib import Path
import matplotlib.pyplot as plt
%matplotlib inline
warnings.filterwarnings('ignore')
pd.set_option('display.width', 200)

# ▼▼▼ 본인 데이터 폴더 경로로 변경하세요 ▼▼▼
DATA_DIR = Path('분석데이터원본')
if not DATA_DIR.exists():
    DATA_DIR = Path('/tmp/khanes_data')   # (실행 환경용 임시 경로)
print('data dir:', DATA_DIR)
```

## 1. 데이터 불러오기 및 review cohort 구성

19–39세 중 혈압 측정값(SBP/DBP)이 있는 대상자로 review cohort를 구성한다.

```python
cand = ['HE_sbp','HE_dbp','HE_BMI','HE_wc','HE_glu','HE_HbA1c','HE_TG','HE_chol',
        'HE_HDL_st2','HE_LDL_drct','HE_HPfh1','HE_HPfh2','HE_HPfh3','mh_stress',
        'DE1_dg','DI2_dg','DI3_dg','DI4_dg','DI5_dg']
base = ['age','sex']

frames = []
for f in sorted(DATA_DIR.glob('HN*_all.sav')):
    year = 2000 + int(f.name[2:4])
    d, _ = pyreadstat.read_sav(str(f))
    keep = [c for c in base+cand if c in d.columns]
    d = d[keep].copy(); d['year'] = year
    frames.append(d)
df = pd.concat(frames, ignore_index=True)

df['covid_period'] = 'before'
df.loc[df['year'].isin([2020,2021]), 'covid_period'] = 'during'
df.loc[df['year'].isin([2022,2023]), 'covid_period'] = 'after'

rev = df[(df['age']>=19) & (df['age']<=39)].dropna(subset=['HE_sbp','HE_dbp']).copy()
print('Review cohort N =', len(rev))
```

## 2. 결측률 스크리닝 (>30% 변수 삭제)

교수님 지침: **결측률 30% 초과 변수는 결측치 대체의 의미가 낮으므로 삭제.**

```python
rows = []
for c in cand:
    if c not in rev.columns: continue
    miss = rev[c].isnull().mean()*100
    rows.append([c, round(miss,1), 'DROP (>30%)' if miss>30 else 'keep'])
miss_tbl = pd.DataFrame(rows, columns=['variable','missing_%','decision'])
miss_tbl
```

**해석**
- **LDL(HE_LDL_drct) 결측 ≈ 68%** → 전체 기간 분석에서는 **삭제**. (2022년부터 측정 시작, 그 이전 ~90% 결측)
- 그 외 추가 변수(허리둘레, HbA1c, 총콜레스테롤, HDL, 가족력, 스트레스, 당뇨/이상지질 진단)는 결측 ≤30% → 결측치 대체 후 사용 가능.
- DI3/DI4/DI5(뇌졸중·심근경색·협심증 진단)는 24–27%로 경계선 → 기저질환 변수 구성 시 주의.

## 3. 측정 방식 변화 증거

연도별 SBP/DBP 평균, HDL 평균, LDL 결측률을 확인한다.

```python
yr = []
for y in range(2017,2024):
    s = rev[rev['year']==y]
    raw_130_80 = ((s.HE_sbp >= 130) | (s.HE_dbp >= 80)).mean() * 100
    raw_140_90 = ((s.HE_sbp >= 140) | (s.HE_dbp >= 90)).mean() * 100
    yr.append({'year':y, 'N':len(s),
               'SBP_mean':round(s.HE_sbp.mean(),1),
               'DBP_mean':round(s.HE_dbp.mean(),1),
               'raw_130_80_%':round(raw_130_80,1),
               'raw_140_90_%':round(raw_140_90,1),
               'HDL_mean':round(s.HE_HDL_st2.mean(),1),
               'LDL_missing_%':round(s.HE_LDL_drct.isnull().mean()*100),
               'BP_missing_%':round(s.HE_sbp.isnull().mean()*100,1)})
year_tbl = pd.DataFrame(yr)
year_tbl
```

```python
fig, ax = plt.subplots(1,2, figsize=(11,3.6))
ax[0].plot(year_tbl.year, year_tbl.SBP_mean, 'o-', label='SBP')
ax[0].plot(year_tbl.year, year_tbl.DBP_mean, 's-', label='DBP')
ax[0].axvline(2020.5, ls='--', c='red', alpha=.6)
ax[0].set_title('Mean SBP / DBP by year'); ax[0].set_xlabel('year'); ax[0].set_ylabel('mmHg'); ax[0].legend(); ax[0].grid(alpha=.3)
ax[1].plot(year_tbl.year, year_tbl.HDL_mean, '^-', c='green', label='HDL mean')
ax[1].axvline(2021.5, ls='--', c='red', alpha=.6)
ax[1].set_title('Mean HDL by year'); ax[1].set_xlabel('year'); ax[1].set_ylabel('mg/dL'); ax[1].legend(); ax[1].grid(alpha=.3)
plt.tight_layout(); plt.show()
```

**해석 — 측정 방식 변화로 판단하는 근거**
- **DBP가 2021년부터 약 4 mmHg 급락**, **SBP는 변화 없음** (빨간 점선 = 2020/2021 경계).
- 이 패턴은 모든 성별·연령 하위군에서 동일하게 나타남(실제 변화라면 SBP도 함께 변해야 함).
- **HDL은 2022년부터 약 +5 상승**, **LDL은 2022년부터 측정 시작**(이전 ~90% 결측).
- KNHANES 혈압 측정 장비가 **2020년 이후 단계적으로 변경, 2023년부터 Microlife로 통일**된 점과 일치하므로, 조사 시기 효과와 측정 장비/환경 변화가 서로 분리되기 어렵다.

## 4. 혈압 3-way 분석 (교수님 지침)

① 원자료 / ② 측정기기 변화 보정 / ③ 측정 일관 연도 제한 — 세 가지를 함께 제시한다.
(130/80 = 2017 ACC/AHA 기준, 140/90 = 기존 국내 기준 비교)

과제3에서는 130/80을 주분석 target으로 사용하고, 140/90은 이전 기준과의 비교 목적으로만 제시한다. 측정 일관 연도 제한분석에서는 2021년을 장비 전환 시기로 보고 구장비/신장비 비교에서 제외하였다.

```python
# batch offset: 구(2017-2020) - 신(2021-2023) DBP 평균차
off = rev[rev.year.isin([2017,2018,2019,2020])].HE_dbp.mean() - rev[rev.year.isin([2021,2022,2023])].HE_dbp.mean()
rev['DBP_adj'] = rev.HE_dbp + rev.year.isin([2021,2022,2023]) * off
print(f'DBP batch offset (old - new) = {off:.2f} mmHg')

def prev(s, dbp_col, sbp_t, dbp_t):
    return round(((s.HE_sbp>=sbp_t) | (s[dbp_col]>=dbp_t)).mean()*100, 1)

# ① 원자료 / ② 보정  (시기별)
res = []
for p in ['before','during','after']:
    s = rev[rev.covid_period==p]
    res.append({'period':p, 'N':len(s),
                '①raw 130/80':prev(s,'HE_dbp',130,80),
                '②adj 130/80':prev(s,'DBP_adj',130,80),
                '①raw 140/90':prev(s,'HE_dbp',140,90),
                '②adj 140/90':prev(s,'DBP_adj',140,90)})
three_way = pd.DataFrame(res)
three_way
```

```python
# ③ 측정 일관 연도 제한분석
old = rev[rev.year<=2020]; new = rev[rev.year>=2022]
restrict = pd.DataFrame([
    {'window':'2017-2020 (old device)', 'N':len(old), '130/80':prev(old,'HE_dbp',130,80), '140/90':prev(old,'HE_dbp',140,90)},
    {'window':'2022-2023 (new device)', 'N':len(new), '130/80':prev(new,'HE_dbp',130,80), '140/90':prev(new,'HE_dbp',140,90)},
])
restrict
```

```python
# 시각화: 시기별 130/80 — 원자료 vs 보정
fig, ax = plt.subplots(figsize=(6,3.6))
x = np.arange(3); w=0.35
ax.bar(x-w/2, three_way['①raw 130/80'], w, label='① raw')
ax.bar(x+w/2, three_way['②adj 130/80'], w, label='② batch-adjusted')
ax.set_xticks(x); ax.set_xticklabels(three_way['period'])
ax.set_ylabel('Hypertension % (130/80)'); ax.set_title('130/80 prevalence: raw vs batch-adjusted')
ax.legend(); ax.grid(axis='y', alpha=.3)
for i,v in enumerate(three_way['①raw 130/80']): ax.text(i-w/2, v+0.4, f'{v}', ha='center', fontsize=8)
for i,v in enumerate(three_way['②adj 130/80']): ax.text(i+w/2, v+0.4, f'{v}', ha='center', fontsize=8)
plt.tight_layout(); plt.show()
```

**해석**
- 원자료 기준: before 26.4% → during 24.0% → after 15.1% 로 감소.
- DBP +4mmHg 민감도 보정 후: 26.4% → 27.4% → 23.0% 로 시기 차이가 줄어듦.
- 구장비(2017–2020) 27.3% vs 신장비(2022–2023) 15.1% → 차이는 장비 차이로 설명될 가능성이 있음.
- ⚠️ ②의 단순 보정(+offset)은 **주분석이 아니라 민감도분석**으로 제시한다.
- 2021년은 장비 전환 시기로 볼 수 있어 측정 일관 연도 제한분석에서 제외하였다.

## 5. HDL / LDL 처리 결정 (교수님 지침)

```python
hdl_old = rev[rev.year<=2021].HE_HDL_st2.mean(); hdl_new = rev[rev.year>=2022].HE_HDL_st2.mean()
print(f'HDL  결측 {rev.HE_HDL_st2.isnull().mean()*100:.1f}%  | 2017-21 평균 {hdl_old:.1f} -> 2022-23 평균 {hdl_new:.1f}  (측정 시기별 변화 +{hdl_new-hdl_old:.1f})')
print(f'LDL  전체 결측 {rev.HE_LDL_drct.isnull().mean()*100:.1f}%  | 2022-23 결측 {new.HE_LDL_drct.isnull().mean()*100:.1f}%')
print()
print('결정:')
print('  HDL → 결측은 낮지만 2022년 이후 측정 시기별 변화 존재 → 주분석 제외 / 민감도분석으로만 사용')
print('  LDL → 전체 결측 >30% → 주분석 삭제, 2022-2023 제한분석에서만 사용')
```

## 6. covid_period 해석 제한 & 두 번째 주제 방향

- `covid_period` 는 측정 장비/환경 변화와 분리되지 않으므로 **인과적 효과로 해석하지 않는다.** 조사 시기 + 측정환경 변화가 함께 반영된 변수로만 제한 해석.
- 두 번째 주제는 "COVID 전후 변화"가 아니라 **"KNHANES 측정체계 변화에 따른 심혈관·대사 변수의 불연속성"** 이라는 **방법론적 주제**로 접근.

## 7. 정리 (교수님 피드백 대응)

| 항목 | 처리 |
|---|---|
| 혈압 | ① 원자료 ② 측정기기 변화 보정(민감도) ③ 일관연도(2017–2020 / 2022–2023) 제한분석 병행 |
| 결측 >30% | LDL 삭제 (전체 68%) |
| HDL | 측정 시기별 변화 → 민감도분석으로만 |
| LDL | 2022–2023 제한분석에서만 |
| 그 외 추가변수 | 결측 ≤30% → 결측치 대체 후 사용 |
| covid_period | 인과 해석 불가, 측정환경 대리변수로 제한 |
| 2nd topic | 측정체계 변화에 따른 변수 불연속성 (방법론적 주제) |
